```python
import pandas as pd
```

```python
import yfinance as yf
```

```python
import numpy as np
```

```python
ticker = '^NDX'
bmark = yf.download(ticker, start='2023-01-01', end='2023-03-31')
```

```python
bmark
```

```python
csvfile = '/content/drive/MyDrive/Colab Notebooks/RSItest.csv' # definuji cestu k souboru
```

```python
exportCBT = pd.read_csv(csvfile, index_col="Date")
```

```python
exportCBT
```

```python
exportCBT.plot()
```

```python
exportCBT.index = pd.to_datetime(exportCBT.index)
bmark = yf.download(ticker, exportCBT.index.min(), exportCBT.index.max())
```

```python
bmark
```

```python
exportCBT.index = pd.to_datetime(exportCBT.index)
bmark = yf.download(ticker, exportCBT.index.min(), exportCBT.index.max() + np.timedelta64(1,'D')) # pridam jeden den pomoci knihovny numpy
```

```python
bmark
```

```python
portfolio = pd.concat((exportCBT, bmark["Adj Close"]), axis = 1, ignore_index = False, sort = True) # spojeni tabulek
portfolio.columns = ["RSItest", "^NDX"] # prejmenovani sloupcu
```

```python
portfolio
```

```python
portfolio = portfolio.fillna(method='ffill')
portfolio
```

```python
portfolio.plot()
```

```python
portfolio["RSI_pctch"] = portfolio["RSItest"].pct_change()
portfolio["NDX_pctch"] = portfolio["^NDX"].pct_change()
portfolio[['RSI_pctch', 'NDX_pctch']].plot()
```

```python
portfolio["RSI_cumprod"] = ((1+portfolio["RSI_pctch"]).cumprod()-1)
portfolio["NDX_cumprod"] = ((1+portfolio["NDX_pctch"]).cumprod()-1)
```

```python
portfolio[['RSI_cumprod', 'NDX_cumprod']].plot()
```
